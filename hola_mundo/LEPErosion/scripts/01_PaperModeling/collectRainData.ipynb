## Create HDF5 file from mat files

```julia
using MAT
using HDF5

folderRaw = "../data/rawData/";
folderPre = "../data/PreData/";
files = filter(x -> endswith(x,".mat"),readdir(folderRaw));

println(length(files))
println(files);


for file in files
    vars = matread(joinpath(folderRaw, file));
    value = vars["WFDB_Shareable"]["MainSRpm"]["value"][:, 1] .* (2 * π / 60) ;
    time = vars["WFDB_Shareable"]["MainSRpm"]["time"][:, 1] ;
    aux = split(file,"_");
    h5write(joinpath(folderPre, "predata.h5"),(aux[1]*"/"*aux[2][1:end-4]*"/AngularSpeed/value"),value);
    h5write(joinpath(folderPre, "predata.h5"),(aux[1]*"/"*aux[2][1:end-4]*"/AngularSpeed/time"),time);
end
```

## Check HDF5 file

```julia
using HDF5
folderPre = "../data/PreData/";

f=h5open(joinpath(folderPre, "predata.h5"),"r");

HDF5.show_tree(f);

fields = keys(f)
for field in fields
    blades=keys(f[field])
    for blade in blades
        print(field*" / "*blade*"\n")

        println(f[field][blade]["LT"][:])
        println(f[field][blade]["R"][:])
    end
end

close(f)
```
