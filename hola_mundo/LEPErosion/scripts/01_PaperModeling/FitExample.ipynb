# FitExample

## Load Packages

```julia
include("../../src/ModuleLEPEE.jl")
using .ModuleLEPEE

using Plots;
```

## Load Data

```julia
R=[39, 41, 43.0];
LT=[6, 3.5, 1.5];

fig_points = plot()
plot!(fig_points, R, LT, seriestype = :scatter, label = "Data points", legend = :topleft)
```

## Fitting Curve

```julia
fitting = fit_curve(LT,R)
```

```julia
A=fitting.minimizer[1]
B=fitting.minimizer[2]
println("A = ", A)
println("B = ", B)

fig_fit = plot(title = "Lifetime", xlabel = "R", ylabel = "LT")
plot!(fig_fit, R, LT, seriestype = :scatter, label = "Data points", legend = :topright)

x = collect(35:0.1:50)
plot!(fig_fit, x, A .* (x .^ B), label = "Fit")
display(fig_fit)
```
