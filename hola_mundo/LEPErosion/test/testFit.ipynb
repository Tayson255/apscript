## Fitting



```julia
println("Denmark");
fit_Denmark = fit_curve(LT_Denmark, R_Denmark);
Denmark = (A=fit_Denmark.minimizer[1], B= fit_Denmark.minimizer[2]);
println("LT = ", Denmark.A, " R ^ ", Denmark.B);
println("A = ", Denmark.A);
println("B = ", Denmark.B);
coeffDenmark = coeffs(fit_Denmark, dsdbestDenmark, windTSRDenmark);
println(coeffDenmark);

#=
println("Sweden");
fit_Sweden = fit_curve(LT_Sweden, R_Sweden);
A_Sweden = fit_Sweden.minimizer[1];
B_Sweden = fit_Sweden.minimizer[2];
println("LT = ", A_Sweden, " R ^ ", B_Sweden);
println("A = ", A_Sweden);
println("B = ", B_Sweden);
coeffSweden = coeffs(fit_Sweden, dsdbestSweden, windTSRSweden);
println(coeffSweden);
=#

```

## Plot fitting 


```julia
fig_fit=plot(ylims=(0,20))
scatter!(fig_fit,R_Denmark,LT_Denmark,label="Experimental Points")
r_sample=1:0.5:46.5;
t_sample_Denmark=Denmark.A .* (r_sample .^ Denmark.B );
t_sample_Denmark=clamp.(t_sample_Denmark,0,20);
plot!(fig_fit,r_sample,t_sample_Denmark, label="Fitted Curve")
```

## Prediction

```julia
SwedenPred = LTPrediction(coeffDenmark,dsdbestSweden,windTSRSweden)
println(SwedenPred);
GlasgowPred = LTPrediction(coeffDenmark,dsdbestGlasgow,windTSRGlasgow)
println(GlasgowPred);
GlasgowOffshorePred = LTPrediction(coeffDenmark,dsdOffshoreGlasgow,windTSRGlasgow)
println(GlasgowOffshorePred);

fig_pred=plot(ylims=(0,20))
scatter!(fig_pred,R_Denmark,LT_Denmark,label="Denmark Experimental Points")
r_sample=1:0.5:46.5;
t_sample_Denmark=Denmark.A .* (r_sample .^ Denmark.B );
t_sample_Denmark=clamp.(t_sample_Denmark,0,20);
plot!(fig_pred,r_sample,t_sample_Denmark, label="Fitted Curve")

scatter!(fig_pred,R_Sweden,LT_Sweden,label="Sweden Experimental Points")
r_sample=1:0.5:46.5;
t_sample_Sweden=SwedenPred.A .* (r_sample .^ SwedenPred.B );
t_sample_Sweden=clamp.(t_sample_Sweden,0,20);
plot!(fig_pred,r_sample,t_sample_Sweden, label="Predicted Curve Sweden")

scatter!(fig_pred,R_Glasgow,LT_Glasgow,label="Glasgow Experimental Points")
r_sample=1:0.5:46.5;
t_sample_Glasgow=GlasgowPred.A .* (r_sample .^ GlasgowPred.B );
t_sample_Glasgow=clamp.(t_sample_Glasgow,0,20);
plot!(fig_pred,r_sample,t_sample_Glasgow, label="Predicted Curve Glasgow")

r_sample=1:0.5:46.5;
t_sample_GlasgowOffshore=GlasgowOffshorePred.A .* (r_sample .^ GlasgowOffshorePred.B );
t_sample_GlasgowOffshore=clamp.(t_sample_GlasgowOffshore,0,20);
plot!(fig_pred,r_sample,t_sample_GlasgowOffshore, label="Predicted Curve GlasgowOffshore")
```
